# Lap Time di Spa e Monza per ogni configurazione

## Monza

```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import numpy as np
from PIL import Image
import glob
#%matplotlib inline
```

```python
# Dati
x = ['Pure Pursuit', 'Safe MPC', 'Fast MPC', 'High Performance MPC']
monza_times = [57.4, 52.35, 50.6, 49.6]

# Colori personalizzati
colors = ['#ca0020', '#f4a582', '#92c5de', '#0571b0'] #['#d62728', '#1f77b4', '#ff7f0e', '#2ca02c']

# Creazione del grafico a barre
plt.figure(figsize=(12, 8))
# Creazione delle barre individualmente con etichette specifiche
bars = []
for i in range(len(x)):
    bar = plt.bar(x[i], monza_times[i], color=colors[i], width=0.7, label=x[i])
    bars.append(bar)

# Aggiunta di etichette sugli assi e titolo
plt.xlabel('Configurazioni', fontsize=18)
plt.ylabel('Lap time (s)', fontsize=18)
#plt.title('Lap time per ciascuna configurazione di guida a Monza', fontsize=18)

# Aggiunta di annotazioni per i valori delle barre
for bar in bars:
    yval = bar[0].get_height()
    plt.text(bar[0].get_x() + bar[0].get_width()/2.0, yval, f'{yval:.1f}s', va='bottom', ha='center', fontsize=18)


# Personalizzazione dei ticks sull'asse Y
plt.yticks(np.arange(0, max(monza_times) + 5, 5), fontsize=16)
plt.ylim(40, 60)
plt.xticks(fontsize=16)

# Aggiunta di una griglia
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.legend(title='Metodi', fontsize=16, title_fontsize='18')

plt.savefig('MPC and PP/monza_lap_time_comparisons.png', dpi=300, bbox_inches='tight')
# Mostra il grafico
plt.show()
```
